```python
set1={1,2,3,4}
set2={3,4,5,6}
print("union",set1|set2)
print("intersection",set1&set2)
print("diffrance",set1-set2)
print("diffrance",set2-set1)
print("dissimiliar",set1^set2)
```

```python
all_students = {"A", "B", "C", "D", "E"}
present_students = {"A", "C", "E"}

absent_students = all_students - present_students

print("All Students:", all_students)
print("Present Students:", present_students)
print("Absent Students:", absent_students)
```

```python
my_courses = {"CS101", "MA201", "PH305", "CH101"}
print(f"Original courses: {my_courses}")
my_courses.remove("MA201")
print(f"After removing 'MA201' with remove(): {my_courses}")
print(f"Courses after attempting to remove non-existent 'PY402': {my_courses}")
my_courses.discard("CH101")
print(f"After discarding 'CH101' with discard(): {my_courses}")
my_courses.discard("EE300")
print(f"Courses after attempting to discard non-existent 'EE300': {my_courses}")
```
